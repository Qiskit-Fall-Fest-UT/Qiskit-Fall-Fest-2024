### ã‚¨ãƒ³ã‚¿ãƒ³ã‚°ãƒ«ãƒ¡ãƒ³ãƒˆ


```python
# ï¼’é‡å­ãƒ“ãƒƒãƒˆå›è·¯ã‚’ä½œæˆã—ã¾ã™ã€‚
qc = QuantumCircuit(2, 2)  # 2é‡å­ãƒ“ãƒƒãƒˆ, 2å¤å…¸ãƒ“ãƒƒãƒˆãƒ¬ã‚¸ã‚¹ã‚¿ãƒ¼

# ã‚²ãƒ¼ãƒˆã‚’é©ç”¨ã—ã¾ã™ã€‚
qc.h(0)
qc.cx(0, 1)  # åˆ¶å¾¡NOTã‚²ãƒ¼ãƒˆ

# æ¸¬å®šã‚²ãƒ¼ãƒˆã‚’è¿½åŠ 
qc.measure(0, 0)  # é‡å­ãƒ“ãƒƒãƒˆq0ã‚’æ¸¬å®šã—ã¦ã€å¤å…¸ãƒ¬ã‚¸ã‚¹ã‚¿ãƒ¼c0ã«å…¥ã‚Œã¾ã™
qc.measure(1, 1)  # é‡å­ãƒ“ãƒƒãƒˆq1ã‚’æ¸¬å®šã—ã¦ã€å¤å…¸ãƒ¬ã‚¸ã‚¹ã‚¿ãƒ¼c1ã«å…¥ã‚Œã¾ã™

# å›è·¯ã‚’æç”»
qc.draw(output="mpl")
```

```python
# ã‚·ãƒŸãƒ¥ãƒ¬ãƒ¼ã‚¿ãƒ¼ã§å®Ÿé¨“
backend = AerSimulator()
sampler = SamplerV2(backend)
job = sampler.run([qc])
result = job.result()

#  æ¸¬å®šã•ã‚ŒãŸå›æ•°ã‚’è¡¨ç¤º
counts = result[0].data.c.get_counts()
print(counts)

# ãƒ’ã‚¹ãƒˆã‚°ãƒ©ãƒ ã§æ¸¬å®šã•ã‚ŒãŸç¢ºç‡ã‚’ãƒ—ãƒ­ãƒƒãƒˆ
plot_histogram(counts)
```

```python
# åˆã‚ã¦å®Ÿè¡Œã™ã‚‹å ´åˆã¯ã€èªè¨¼æƒ…å ±ã‚’ãƒ‡ã‚£ã‚¹ã‚¯ã«ä¿å­˜ã—ã¾ã™ï¼š
# https://quantum.ibm.com/ ã®API Tokenã‚’Colabã®å·¦ã‚µã‚¤ãƒ‰ãƒãƒ¼ã®ğŸ”‘ã‚¿ãƒ–ã‹ã‚‰å…¥åŠ›ã—ã¾ã™
from google.colab import userdata

if len(QiskitRuntimeService.saved_accounts())==0:
    QiskitRuntimeService.save_account(channel='ibm_quantum', instance='ibm-q/open/main', token=userdata.get("ibmToken"))

service = QiskitRuntimeService()
service.backends()
```

```python
# ä»¥ä¸‹ã§ãƒ‡ãƒã‚¤ã‚¹ã‚’æŒ‡å®šã§ãã¾ã™ã€‚
backend = service.backend("ibm_kyoto")  # ä¾‹ï¼‰ibm_kyotoã‚’ä½¿ã†å ´åˆ
```

```python
# ä¸€ç•ªç©ºã„ã¦ã„ã‚‹ãƒãƒƒã‚¯ã‚¨ãƒ³ãƒ‰ã‚’è‡ªå‹•çš„ã«é¸æŠã™ã‚‹ã“ã¨ã‚‚ã§ãã¾ã™
backend = service.least_busy(operational=True)
print("æœ€ã‚‚ç©ºã„ã¦ã„ã‚‹ãƒãƒƒã‚¯ã‚¨ãƒ³ãƒ‰ã¯: ", backend)
```

```python
# å®Ÿæ©Ÿã®ãƒãƒƒã‚¯ã‚¨ãƒ³ãƒ‰ã§ã®å®Ÿè¡Œã«æœ€é©ãªå›è·¯ã«å¤‰æ›ã—ã¾ã™
pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
isa_circuit = pm.run(qc)
isa_circuit.draw("mpl", idle_wires=False)
```

```python
# Samplerã§å®Ÿè¡Œã—ã¾ã™
sampler = SamplerV2(backend)
job = sampler.run([isa_circuit])

print("job id:", job.job_id())  # å®Ÿè¡Œã«æ™‚é–“ãŒã‹ã‹ã‚‹ã®ã§job_idã‚’è¡¨ç¤ºã—ã¾ã™
```
