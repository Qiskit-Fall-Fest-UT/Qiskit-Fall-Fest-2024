##  Qiskit pattern step 1: 回路と演算子をマッピングする <a name="step-1"></a>

問題を回路と演算子にマッピングする作業は、少数の量子ビットを扱う場合には比較的簡単ですが、スケールアップすると難しくなることがあります。まずは、ベル状態を構築してみましょう。これは、2つの量子ビット間のエンタングルメント（量子もつれ）を示すクラシックな（*あるいは「量子的な」と言うべきでしょうか*）例です。

最初に、量子ビット0にハダマードゲートを適用し、スーパー ポジションの状態にします。その後、量子ビット0と1の間にCNOTゲートを適用します。このCNOTゲートは量子ビットをエンタングル状態にするため、回路の出力を測定すると、両方の量子ビットが同じ測定結果を共有することが期待されます。

これが理解できたなら、素晴らしいです。準備は整っています。一方で、もしこの内容が難しいと感じた場合、採点対象のノートブックはかなり難しく感じるかもしれません。この機会に、[量子情報と計算](https://learning.quantum.ibm.com/course/basics-of-quantum-information/quantum-circuits)のコースを見直すか、Qiskit Fall Festイベントのワークショップに参加することを検討してください。

<!--
Mapping your problem to circuits and operators is fairly easy when you're only dealing with a handful of qubits, but can get more difficult as you scale up. To start, we will construct a Bell state. This is a classic (*or should I say, "quantum"*) example that showcases entanglement between two qubits.

First we apply a Hadamard gate to qubit 0, placing it in a superposition. Then we apply a CNOT gate between qubits 0 and 1. The CNOT effectively entangles our qubits, meaning that when we measure the output of this circuit we should expect to see that both qubits share a measurement result.

```python
# Create a new circuit with two qubits
qc = QuantumCircuit(2)

# Add a Hadamard gate to qubit 0
qc.h(0)

# Perform a CNOT gate on qubit 1, controlled by qubit 0
qc.cx(0, 1)

# Return a drawing of the circuit using MatPlotLib ("mpl"). This is the
# last line of the cell, so the drawing appears in the cell output.
qc.draw("mpl")
```

私たちは、回路が期待通りに動作していることを視覚化できるようにしたいと考えています。そのためには、いくつかの**演算子**を設定する必要があります。

演習2では、複数の演算子が作成されるのを見ることができます。一部の演算子（例えば`XX`）は両方の量子ビットを測定しますが、他の演算子（例えば`XI`）は片方のみを測定します。後ほど、このノートブックではこれらの演算子の期待値を確認し、ベル状態の回路が期待通りエンタングルされていることを確認します。

エンタングルされたベル状態を作成したため、片方の量子ビットのみを測定する演算子の期待値は0であることが期待されます。同様に、両方の量子ビットを測定する演算子は、ゼロ以外の期待値（理想的には1に近い値）を返すことが期待されます。

では、これらの演算子をどのように作成するのでしょうか？`X`、`Z`、そしてIdentity（`I`）ゲートの組み合わせを使用します。

* 量子ビットに`I`ゲートを適用することは、何も操作しないのと同じです。`X`/`Z`と`I`を組み合わせることで、`XI`、`IX`、`ZI`、`IZ`の演算子を作成できます。
* 両方の量子ビットを測定する演算子は、`XX`と`ZZ`の2つだけです。それぞれ`X`または`Z`ゲートの組み合わせを使用して作成します。

後ほど、このノートブックではこれらの演算子の期待値をグラフ化し、回路を視覚化します。

回路を最適化するためには、Qiskitトランスパイラーを使用できます。トランスパイルの過程では、例えば、必要なゲートの数を最小化するために量子ビットを最適に配置することが可能です。このプロセスの詳細は、採点対象ノートブック3で学ぶことができます。今は、パターンの次のステップに進んでください。

## Qiskit pattern step 4: 結果の後処理 <a name="step-4"></a>

Qiskitパターンの最後のステップでは、ジョブの結果を処理、分析、そして視覚化します。

<!-- The last step of Qiskit patterns is where we process, analyze, and visualize the results of our job.

Here we are going to collect our data and plot it on a graph we construct. -->

```python
# Collect the data
data = ['IZ', 'IX', 'ZI', 'XI', 'ZZ', 'XX']
values = job.result()[0].data.evs

# Set up our graph
container = plt.plot(data, values, '-o')

# Label each axis
plt.xlabel('Observables')
plt.ylabel('Values')

# Draw the final graph
plt.show()
```

<!-- You *should* see the operators which measure both qubits (`ZZ` and `XX`) have a value that's close to 1, and the other operators have a value that's close to 0. This shows that your circuit is entangled. Great work.
